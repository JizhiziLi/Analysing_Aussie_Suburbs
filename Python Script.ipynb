```python
#Please add the Suburbs folder in same folder as this script 
import pandas as pd
%matplotlib inline

filenames = ['Ascot-Vale-Suburb - XLSX.xlsx',
'Braybrook-Suburb - XLSX.xlsx', 'Craigieburn-Suburb - XLSX.xlsx',
'Croydon-Suburb - XLSX.xlsx', 'Fawkner-Suburb - XLSX.xlsx',
'Footscray-Suburb - XLSX.xlsx','Glenroy-Suburb - XLSX.xlsx', 
'Malvern-East-Suburb - XLSX.xlsx', 'Malvern-Suburb - XLSX.xlsx',
'Melbourne-Airport-Suburb - XLSX.xlsx', 'Mentone-Suburb - XLSX.xlsx',
'Moorabbin-Suburb - XLSX.xlsx', 'Mordialloc-Suburb - XLSX.xlsx', 
'Murrumbeena-Suburb - XLSX.xlsx', 'Noble-Park-Suburb - XLSX.xlsx',
'North-Melbourne-Suburb - XLSX.xlsx', 'Northcote-Suburb - XLSX.xlsx',
'Parkville-Suburb - XLSX.xlsx', 'Pascoe-Vale-South-Suburb - XLSX.xlsx',
'Port-Melbourne-Suburb - XLSX.xlsx','Prahran-Suburb - XLSX.xlsx',
'Somerville-Suburb - XLSX.xlsx','Sorrento-Suburb - XLSX.xlsx', 
'South-Melbourne-Suburb - XLSX.xlsx','South-Yarra-Suburb - XLSX.xlsx',
'Springvale-Suburb - XLSX.xlsx','St-Andrews-Beach-Suburb - XLSX.xlsx',
'St-Kilda-East-Suburb - XLSX.xlsx','St-Kilda-Suburb - XLSX.xlsx',
'St-Kilda-West-Suburb - XLSX.xlsx','Toorak-Suburb - XLSX.xlsx',
'Tyabb-Suburb - XLSX.xlsx','Waterways-Suburb - XLSX.xlsx',
'Windsor-Suburb - XLSX.xlsx']
#Add data to data frames and merge together
frames=[]
for file in filenames:
    xl_file = pd.read_excel("Suburbs/"+ file,sheetname='data',index_col='Community Name',parse_cols=[1,2])
    xl_file = xl_file.transpose()
    frames.append(xl_file)
df = pd.concat(frames)
df
```

```python
#Clustering of data based on info retrieved in the previous cell
%matplotlib inline
import pandas
import matplotlib
import numpy
import matplotlib.pyplot as plt
from pandas import *
from scipy.spatial.distance import cosine
from sklearn import manifold
from sklearn.metrics import euclidean_distances
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

suburbs = ['Ascot-Vale',
'Braybrook', 'Craigieburn',
'Croydon', 'Fawkner',
'Footscray','Glenroy', 
'Malvern-East', 'Malvern',
'Melbourne-Airport', 'Mentone',
'Moorabbin', 'Mordialloc', 
'Murrumbeena', 'Noble-Park',
'North-Melbourne', 'Northcote',
'Parkville', 'Pascoe-Vale-South',
'Port-Melbourne','Prahran',
'Somerville','Sorrento', 
'South-Melbourne','South-Yarra',
'Springvale','St-Andrews-Beach',
'St-Kilda-East','St-Kilda',
'St-Kilda-West','Toorak',
'Tyabb','Waterways',
'Windsor']

def cluster_data(data, filename):
    X_std = StandardScaler().fit_transform(data)
    est = KMeans(4)  # 4 clusters
    est.fit(X_std)
    y_kmeans = est.predict(X_std)
    clust = Series(y_kmeans)
    clustlabel = Series(suburbs)
    newdata=concat([clustlabel,clust], axis=1)
    newdata
    newdata.columns=("suburbs","cluster")

    for i in range(4):
        cl1 = newdata[newdata.cluster==i]
        with open(filename, 'a') as f:
             cl1.to_csv(f, header=False)
        print (cl1)
    f.close()
    
    Xnew = PCA(2).fit_transform(X_std)

    
    plt.figure(figsize=(20,20))
    font = {'family' : 'normal',
                'size'   : 15}

    matplotlib.rc('font', **font)
    plt.scatter(Xnew[:, 0], Xnew[:, 1], c=y_kmeans, s=200, cmap='rainbow')
    for label, x, y in zip(suburbs, Xnew[:, 0], Xnew[:, 1]):
            plt.annotate(
                label,
                xy = (x, y), xytext = (0, 0),
                textcoords = 'offset points', ha = 'left', va = 'top')
    plt.show()

    
comb_feat=read_csv("All_Features.csv")
cluster_data(comb_feat,"clusteranalysis.csv")
```
